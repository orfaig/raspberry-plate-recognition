**×¤×¨××˜×¨×™× ×—×©×•×‘×™×:**

## ğŸ› ï¸ ×©×œ×‘ 3: ×¤×•× ×§×¦×™×•×ª ×¢×–×¨ - Smooth Steering

× ×‘× ×” ×¤×•× ×§×¦×™×” ×©×××¤×©×¨×ª ×‘×§×¨×ª ×”×™×’×•×™ ×—×œ×§×”:

````python
def smooth_steering(base_speed, steering_value, steering_gain=50):
    """
    ××—×©×‘ ××”×™×¨×•×™×•×ª ×’×œ×’×œ×™× ×¢×‘×•×¨ ×”×™×’×•×™ ×—×œ×§
    
    Parameters:
    -----------
    base_speed : int (0-255)
        ××”×™×¨×•×ª ×”×‘×¡×™×¡ - ×§×•×‘×¢×ª ×›××” ××”×™×¨ ×”×¨×•×‘×•×˜ × ×•×¡×¢
        
    steering_value : float (-1.0 to 1.0)
        ×›×™×•×•×Ÿ ×”×¤× ×™×™×”:
        - 0 = ×™×©×¨
        - 1.0 = ×¤× ×™×™×” ×—×“×” ×™××™× ×”
        - -1.0 = ×¤× ×™×™×” ×—×“×” ×©×××œ×”
        - 0.3 = ×¤× ×™×™×” ×¢×“×™× ×” ×™××™× ×”
        
    steering_gain : int (default=50)
        ×¢×•×¦××ª ×”×¤× ×™×™×” - ×›×›×œ ×©×’×‘×•×” ×™×•×ª×¨, ×”×¤× ×™×™×” ×—×“×” ×™×•×ª×¨
        ×¢×¨×›×™× ××•××œ×¦×™×: 30-80
    
    Returns:
    --------
    tuple : (left_speed, right_speed)
        ××”×™×¨×•×™×•×ª ××—×•×©×‘×•×ª ×œ×’×œ×’×œ ×©×××œ ×•×™××™×Ÿ
    
    Examples:
    ---------
    >>> smooth_steering(100, 0, 50)
    (100, 100)  # × ×¡×™×¢×” ×™×©×¨
    
    >>> smooth_steering(100, 0.5, 50)
    (125, 75)  # ×¤× ×™×™×” ×™××™× ×” ×‘×™× ×•× ×™×ª
    
    >>> smooth_steering(100, -1.0, 60)
    (40, 160)  # ×¤× ×™×™×” ×—×“×” ×©×××œ×”
    """
    # ×—×™×©×•×‘ ×ª×™×§×•×Ÿ ×”××”×™×¨×•×ª ×œ×›×œ ×’×œ×’×œ
    correction = steering_value * steering_gain
    
    # ×’×œ×’×œ ×©×××œ: ××”×™×¨ ×™×•×ª×¨ ×‘×¤× ×™×™×” ×™××™× ×”
    left_speed = base_speed + correction
    
    # ×’×œ×’×œ ×™××™×Ÿ: ××”×™×¨ ×™×•×ª×¨ ×‘×¤× ×™×™×” ×©×××œ×”
    right_speed = base_speed - correction
    
    # ×”×’×‘×œ×” ×œ×˜×•×•×— ×—×•×§×™ (0-255)
    left_speed = max(0, min(255, int(left_speed)))
    right_speed = max(0, min(255, int(right_speed)))
    
    return left_speed, right_speed


def drive_with_steering(base_speed, steering_value, duration, steering_gain=50):
    """
    × ×•×¡×¢ ×¢× ×”×™×’×•×™ ×œ××©×š ×–××Ÿ ××¡×•×™×
    
    Parameters:
    -----------
    base_speed : int
        ××”×™×¨×•×ª ×‘×¡×™×¡×™×ª
    steering_value : float
        ×¢×¨×š ×”×™×’×•×™ (-1.0 ×¢×“ 1.0)
    duration : float
        ×–××Ÿ × ×¡×™×¢×” ×‘×©× ×™×•×ª
    steering_gain : int
        ×¢×•×¦××ª ×”×™×’×•×™
    """
    left, right = smooth_steering(base_speed, steering_value, steering_gain)
    print(f"ğŸš— × ×•×¡×¢: L={left}, R={right}, steering={steering_value:.2f}, ×–××Ÿ={duration}s")
    car.Car_Run(left, right)
    time.sleep(duration)
    car.Car_Stop()


print("âœ… ×¤×•× ×§×¦×™×•×ª ×”×™×’×•×™ ××•×›× ×•×ª!")
````

### ×ª×¨×—×™×©: ×”×¨×•×‘×•×˜ ×¢×•×§×‘ ××—×¨×™ QR Code

×‘××¦×œ××”, ×”-QR ×™×›×•×œ ×œ×”×™×•×ª:
- **×‘××¨×›×– ×”××¡×š** â†’ steering = 0 (×™×©×¨)
- **××©×××œ ×œ××¡×š** â†’ steering ×©×œ×™×œ×™ (×¤× ×” ×©×××œ×”)
- **××™××™×Ÿ ×œ××¡×š** â†’ steering ×—×™×•×‘×™ (×¤× ×” ×™××™× ×”)

× ×“××” ×ª×¨×—×™×© ×©×‘×• ×”-QR ×–×–:

````python
def simulate_qr_tracking():
    """
    ××“××” ××¢×§×‘ ××—×¨×™ QR ×©×–×–
    
    ×‘×¤×•×¢×œ, ×¢×¨×›×™ ×”-steering ×™×’×™×¢×• ×××™×§×•× ×”-QR ×‘××¦×œ××”:
    steering = (qr_x_position - image_center_x) / image_width
    
    ×œ×“×•×’××”:
    - QR ×‘××¨×›×– (x=320, ××¨×›×–=320) â†’ steering = 0
    - QR ×‘×©×××œ (x=200, ××¨×›×–=320) â†’ steering = -0.375 (×¤× ×” ×©×××œ×”)
    - QR ×‘×™××™×Ÿ (x=440, ××¨×›×–=320) â†’ steering = +0.375 (×¤× ×” ×™××™× ×”)
    """
    print("\nğŸ¯ ××ª×—×™×œ ×¡×™××•×œ×¦×™×” - ××¢×§×‘ QR\n")
    print("(×‘×¤×•×¢×œ, ×”×¢×¨×›×™× ×™×—×•×©×‘×• ×××™×§×•× ×”-QR ×‘××¦×œ××”)\n")
    
    # ×¡×™××•×œ×¦×™×”: ×”-QR "×–×–" ×‘××¡×š
    simulated_qr_positions = [
        ("QR ×‘××¨×›×–", 0),
        ("QR ×–×– ×™××™× ×” ××¢×˜", 0.2),
        ("QR ×™××™× ×”", 0.4),
        ("QR ×™××™× ×” ×××•×“", 0.6),
        ("QR ×—×•×–×¨ ×œ××¨×›×–", 0.3),
        ("QR ×‘××¨×›×–", 0),
        ("QR ×–×– ×©×××œ×” ××¢×˜", -0.2),
        ("QR ×©×××œ×”", -0.4),
        ("QR ×©×××œ×” ×××•×“", -0.6),
        ("QR ×—×•×–×¨ ×œ××¨×›×–", -0.3),
        ("QR ×‘××¨×›×– - ××¦×•×™×Ÿ!", 0)
    ]
    
    for position_desc, steering in simulated_qr_positions:
        print(f"ğŸ“ {position_desc} (steering={steering:+0.1f})")
        drive_with_steering(
            base_speed=70,
            steering_value=steering,
            duration=0.7,
            steering_gain=60
        )
        time.sleep(0.2)
    
    print("\nâœ… ×”×¨×•×‘×•×˜ ×¢×§×‘ ×‘×”×¦×œ×—×” ××—×¨×™ ×”-QR!")
    print("ğŸ’¡ ×‘×¤×¨×•×™×§×˜ 4 × ×—×‘×¨ ××ª ×–×” ×œ××¦×œ××” ×××™×ª×™×ª")

# ×”×¨×¥ ×¡×™××•×œ×¦×™×”
simulate_qr_tracking()
````


````python
def steering_calculator(qr_x, image_width=640, base_speed=100, steering_gain=50):
    """
    ××—×©×‘×•×Ÿ ×œ×—×™×©×•×‘ steering ×•××”×™×¨×•×™×•×ª ×’×œ×’×œ×™×
    
    Parameters:
    -----------
    qr_x : int
        ××™×§×•× ×”-QR ×‘×¦×™×¨ X (0 = ×©×××œ ×”××¡×š, 640 = ×™××™×Ÿ)
    image_width : int
        ×¨×•×—×‘ ×”×ª××•× ×” ×‘×¤×™×§×¡×œ×™×
    base_speed : int
        ××”×™×¨×•×ª ×‘×¡×™×¡×™×ª
    steering_gain : int
        ×¢×•×¦××ª ×”×™×’×•×™
    """
    center_x = image_width / 2
    error = qr_x - center_x
    steering_value = error / center_x
    
    # ×—×™×©×•×‘ ××”×™×¨×•×™×•×ª
    left, right = smooth_steering(base_speed, steering_value, steering_gain)
    
    # ×”×¦×’×ª ×ª×•×¦××•×ª
    print("\n" + "="*60)
    print("ğŸ§® ××—×©×‘×•×Ÿ Steering")
    print("="*60)
    print(f"ğŸ“ QR Position: x={qr_x} (××¨×›×–={center_x})")
    print(f"âŒ Error: {error:+.1f} ×¤×™×§×¡×œ×™×")
    print(f"ğŸ¯ Steering Value: {steering_value:+.3f}")
    print(f"\nâš™ï¸ ×¤×¨××˜×¨×™×:")
    print(f"   Base Speed: {base_speed}")
    print(f"   Steering Gain: {steering_gain}")
    print(f"\nğŸš— ××”×™×¨×•×™×•×ª ××—×•×©×‘×•×ª:")
    print(f"   ×’×œ×’×œ ×©×××œ: {left}")
    print(f"   ×’×œ×’×œ ×™××™×Ÿ: {right}")
    print(f"\nğŸ“ ×›×™×•×•×Ÿ: ", end="")
    
    if abs(steering_value) < 0.1:
        print("â¬†ï¸ ×™×©×¨ ×§×“×™××”")
    elif steering_value > 0:
        intensity = "×—×“×”" if abs(steering_value) > 0.5 else "×¢×“×™× ×”"
        print(f"â¡ï¸ ×¤× ×™×™×” {intensity} ×™××™× ×”")
    else:
        intensity = "×—×“×”" if abs(steering_value) > 0.5 else "×¢×“×™× ×”"
        print(f"â¬…ï¸ ×¤× ×™×™×” {intensity} ×©×××œ×”")
    print("="*60)

# ×“×•×’×××•×ª ×©×™××•×©
print("\nğŸ“š ×“×•×’×××•×ª:\n")

# QR ×‘××¨×›×–
steering_calculator(qr_x=320, base_speed=100, steering_gain=50)

# QR ×‘×™××™×Ÿ
steering_calculator(qr_x=500, base_speed=100, steering_gain=50)

# QR ×‘×©×××œ
steering_calculator(qr_x=150, base_speed=100, steering_gain=50)
````
