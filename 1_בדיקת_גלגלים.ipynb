זה כמו אופניים - אם רוכב אחד רוכב מהר יותר מהשני, האופניים פונים!

## 🎯 אתגר: מסלול מלבני

נשלב את כל מה שלמדנו ליצירת מסלול מלבן:
1. קדימה
2. פנייה ימינה
3. קדימה
4. פנייה ימינה
5. (וכן הלאה... 4 צלעות)

**טיפ:** אפשר לשנות את זמני הנסיעה והפניות כדי להתאים לחלל שלכם

```python
print("🔲 מתחיל מסלול מלבני...")

for i in range(4):  # 4 צלעות של מלבן
    print(f"צלע {i+1}: נוסע קדימה")
    car.Car_Run(120, 120)
    time.sleep(1.5)  # נסיעה ישר
    
    car.Car_Stop()
    time.sleep(0.3)
    
    print(f"פנייה {i+1}: פונה ימינה")
    car.Car_Run(180, 60)  # פנייה חדה ימינה
    time.sleep(0.8)  # זמן פנייה - נסו לכוונן!
    
    car.Car_Stop()
    time.sleep(0.3)

car.Car_Stop()
print("🏁 סיימתי את המסלול!")
```

## 📊 טבלת פרמטרים - סיכום

| פקודה | פרמטר 1 | פרמטר 2 | תיאור |
|-------|---------|---------|-------|
| `Car_Run(L, R)` | מהירות שמאל (0-255) | מהירות ימין (0-255) | נוסע קדימה |
| `Car_Stop()` | - | - | עוצר לגמרי |
| `time.sleep(T)` | זמן בשניות | - | ממתין T שניות |

### דוגמאות שימושיות:
- **נסיעה ישר:** `Car_Run(100, 100)`
- **פנייה ימינה:** `Car_Run(150, 50)` - שמאל מהיר, ימין איטי
- **פנייה שמאלה:** `Car_Run(50, 150)` - שמאל איטי, ימין מהיר
- **סיבוב במקום (ימינה):** `Car_Run(150, -150)` - גלגל אחד קדימה, השני אחורה
