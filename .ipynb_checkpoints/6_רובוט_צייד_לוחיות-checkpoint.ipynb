## ⚙️ שלב 2: הגדרת כל הפרמטרים

**פרמטרי AI:**
- `CONF_THRES` - סף זיהוי
- `IOU_THRES` - מניעת כפילויות
- `INFER_PERIOD_S` - תדירות AI

**פרמטרי OCR:**
- `OCR_EVERY_N_FRAMES` - כל כמה פריימים
- `MARGIN_X/Y` - שוליים לגזירה

**פרמטרי היגוי:**
- `SPEED` - מהירות בסיס
- `YAW_DEADBAND` - איזור מת
- `STEER_KP` - מקדם היגוי
- `STEER_LIMIT` - גבול תיקון

**פרמטרי לוחית:**
- `MIN_W/H` - גודל מינימלי
- `MIN/MAX_ASPECT` - יחס רוחב/גובה
- `TARGET_RATIO` - תנאי הגעה (0.6 = 60% מהמסך)

**מצלמה:**
- `FX, FY` - focal length
- `CX, CY` - מרכז

**משימה:**
- `TARGET_PLATE` - המספר שאנחנו מחפשים!

````python
# ===== מטרה =====
TARGET_PLATE = "91-179-18"  # שנו לפי הצורך!

# ===== AI =====
ONNX_PATH = "/home/pi/Yahboom_project/uveye/code/raspberry-plate-recognition/yolov8n-license_plate.onnx"
CONF_THRES = 0.2
IOU_THRES = 0.2
INFER_PERIOD_S = 0.10

# ===== OCR =====
OCR_EVERY_N_FRAMES = 3
MARGIN_X = 0.10
MARGIN_Y = 0.25

# ===== היגוי =====
SPEED = 80.0
YAW_DEADBAND = 5.0
STEER_KP = 0.020
STEER_LIMIT = 0.25
YAW_FILTER_ALPHA = 0.30  # סינון רעש
LOST_TIMEOUT_S = 3.0     # זמן המתנה אם איבדנו

# ===== פילטרי לוחית =====
MIN_W = 40
MIN_H = 15
MIN_ASPECT = 2.0
MAX_ASPECT = 7.0
TARGET_RATIO = 0.6

# ===== מצלמה =====
FX = 450.0
FY = 600.0
CX = 320.0
CY = 240.0

K = np.array([[FX, 0.0, CX],
              [0.0, FY, CY],
              [0.0, 0.0, 1.0]], dtype=np.float32)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

print(f"🎯 מטרה: {TARGET_PLATE}")
print(f"🚀 מהירות: {SPEED}")
print(f"🛑 עצירה כשלוחית תופסת {TARGET_RATIO*100:.0f}% מהמסך")
````

## 🔧 שלב 3: פונקציות עזר

נכין את כל הפונקציות:
1. **פילטרים** - האם הלוחית "טובה"?
2. **OCR** - קריאת המספר
3. **היגוי** - חישוב PWM
4. **HUD** - תצוגה על המסך

````python
# פילטר לוחית
def passes_plate_filters(bbox):
    x1, y1, x2, y2 = bbox
    w = x2 - x1
    h = y2 - y1
    aspect = w / max(1, h)
    
    size_ok = (w >= MIN_W) and (h >= MIN_H)
    aspect_ok = (MIN_ASPECT <= aspect <= MAX_ASPECT)
    
    return size_ok and aspect_ok

# נורמליזציה
def normalize_plate(raw):
    digits = re.sub(r"\D", "", raw or "")
    if len(digits) == 7:
        return f"{digits[:2]}-{digits[2:5]}-{digits[5:]}"
    if len(digits) == 8:
        return f"{digits[:3]}-{digits[3:5]}-{digits[5:]}"
    return None

# גזירה
def crop_plate(frame, bbox):
    x1, y1, x2, y2 = bbox
    w, h = x2 - x1, y2 - y1
    mx = int(w * MARGIN_X)
    my = int(h * MARGIN_Y)
    return frame[
        max(0, y1 - my):min(frame.shape[0], y2 + my),
        max(0, x1 - mx):min(frame.shape[1], x2 + mx)
    ]

# OCR
def ocr_plate(crop_img):
    if crop_img is None or crop_img.size == 0:
        return None
    gray = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
    gray = clahe.apply(gray)
    gray = cv2.resize(gray, None, fx=2.4, fy=2.4, interpolation=cv2.INTER_CUBIC)
    gray = cv2.GaussianBlur(gray, (3, 3), 0)
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    for img in (binary, 255 - binary):
        text = pytesseract.image_to_string(img, config=CONFIG_TESS)
        plate = normalize_plate(text)
        if plate:
            return plate
    return None

# היגוי
def compute_steering(yaw):
    if abs(yaw) <= YAW_DEADBAND:
        return SPEED, SPEED
    steer = STEER_KP * yaw
    steer = np.clip(steer, -STEER_LIMIT, STEER_LIMIT)
    left = SPEED * (1.0 + steer)
    right = SPEED * (1.0 - steer)
    return np.clip(left, 0, 100), np.clip(right, 0, 100)

# HUD
def draw_hud(frame, yaw, status, L, R, plate_text=None):
    vis = frame.copy()
    h, w = vis.shape[:2]
    bar_h = 110
    
    overlay = vis.copy()
    cv2.rectangle(overlay, (0, 0), (w, bar_h), (0, 0, 0), -1)
    vis = cv2.addWeighted(overlay, 0.35, vis, 0.65, 0)
    
    # חץ
    cx, cy = w // 2, 75
    arrow_len = 50
    yaw_c = np.clip(yaw, -45, 45)
    theta = np.radians(yaw_c)
    ex = int(cx + arrow_len * np.sin(theta))
    ey = int(cy - arrow_len * np.cos(theta))
    cv2.arrowedLine(vis, (cx, cy), (ex, ey), (0, 255, 0), 4, tipLength=0.3)
    
    # טקסט
    cv2.putText(vis, f"Yaw: {yaw:.1f}", (10, bar_h - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
    
    color = (0, 255, 0) if "TARGET" in status else (255, 255, 255)
    cv2.putText(vis, f"Status: {status}", (10, bar_h + 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)
    
    cv2.putText(vis, f"L:{L:.0f} R:{R:.0f}", (w - 170, bar_h - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)
    
    if plate_text:
        cv2.putText(vis, f"Plate: {plate_text}", (10, bar_h + 55),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 3)
    
    return vis

print("✅ פונקציות עזר מוכנות!")
````

## 🎯 שלב 5: המשימה המלאה! (ללא תנועה - תצוגה בלבד)

לפני שנתן לרובוט לזוז, נבדוק שהכל עובד.

**מה יוצג:**
- ריבועים סביב לוחיות
- חץ שמראה לאן לפנות
- PWM מחושב (אבל לא נזיז)
- טקסט לוחית כשנקרא
- "TARGET FOUND" כשמוצאים את המטרה

````python
# Best-practice smooth live camera in Jupyter: ipywidgets.Image (no cv2.imshow / no matplotlib redraw)
# Stop the stream by interrupting the cell (Kernel -> Interrupt) or the stop button.

import cv2
import time
import ipywidgets as widgets
from IPython.display import display

def free_camera(
    index=0,
    width=640,
    height=480,
    use_v4l2=False,          # True on Raspberry Pi / Linux if needed
    buffersize=1,
    warmup=20,
    jpeg_quality=75
):
    backend = cv2.CAP_V4L2 if use_v4l2 else 0
    cap = cv2.VideoCapture(index, backend) if backend else cv2.VideoCapture(index)

    cap.set(cv2.CAP_PROP_FRAME_WIDTH,  width)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, height)
    cap.set(cv2.CAP_PROP_BUFFERSIZE,   buffersize)  # may be ignored, but helps when supported

    if not cap.isOpened():
        raise RuntimeError("Cannot open camera")

    # warm-up + drop stale frames
    for _ in range(warmup):
        cap.read()

    img = widgets.Image(format="jpeg")
    display(img)

    def show_frame(frame_bgr):
        ok, jpg = cv2.imencode(
            ".jpg",
            frame_bgr,
            [int(cv2.IMWRITE_JPEG_QUALITY), int(jpeg_quality)]
        )
        if ok:
            img.value = jpg.tobytes()

    return cap, show_frame


# --- use it in your existing loop (replace cv2.imshow / waitKey block) ---

cap, show_frame = free_camera(index=0, width=640, height=480, use_v4l2=False, jpeg_quality=75)

print(f"🎯 מחפש: {TARGET_PLATE}")
print("מצב תצוגה (ללא תנועה)\n")

last_infer = 0.0
detections = []
frame_count = 0
target_found = False
target_plate_text = None
yaw_filtered = 0.0

# streaming loop parameters
TARGET_FPS = 20
DT = 1.0 / TARGET_FPS
DROP_GRABS = 2

t_next = time.time()

try:
    for i in range(500):
        # drop buffered frames to reduce lag
        for _ in range(DROP_GRABS):
            cap.grab()

        ret, frame = cap.read()
        if not ret:
            time.sleep(0.02)
            continue

        now = time.time()
        frame_count += 1
        fh, fw = frame.shape[:2]

        # AI
        if now - last_infer >= INFER_PERIOD_S:
            output = run_onnx_inference(frame)
            detections = decode_yolov8(output, fw, fh, CONF_THRES, IOU_THRES)
            last_infer = now

        yaw = 0.0
        L, R = 0, 0
        status = "looking for..."

        if target_found:
            status = "🎉 TARGET FOUND!"
        elif detections:
            best = max(detections, key=lambda d: d["confidence"])
            x1, y1, x2, y2 = best["bbox"]

            if passes_plate_filters((x1, y1, x2, y2)):
                cx, cy = bbox_center((x1, y1, x2, y2))
                yaw, _ = pixel_to_angles(cx, cy, K)
                yaw_filtered = (1 - YAW_FILTER_ALPHA) * yaw_filtered + YAW_FILTER_ALPHA * yaw

                L, R = compute_steering(yaw_filtered)

                bbox_w = x2 - x1
                if bbox_w >= (fw * TARGET_RATIO):
                    status = "Arrived!"
                else:
                    status = f"Drive (w={bbox_w}px)"

                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.circle(frame, (int(cx), int(cy)), 5, (0, 0, 255), -1)

                # OCR
                if not target_found and frame_count % OCR_EVERY_N_FRAMES == 0:
                    crop = crop_plate(frame, (x1, y1, x2, y2))
                    plate = ocr_plate(crop)
                    if plate:
                        print(f"📋 קראתי: {plate}")
                        if plate == TARGET_PLATE:
                            target_found = True
                            target_plate_text = plate
                            print(f"\n🎉🎉🎉 מצאתי את {TARGET_PLATE}! 🎉🎉🎉\n")

        vis = draw_hud(frame, yaw_filtered, status, L, R, target_plate_text)

        # Jupyter display (replaces cv2.imshow + cv2.waitKey)
        show_frame(vis)

        # pace output (prevents CPU spikes)
        now = time.time()
        sleep = t_next - now
        if sleep > 0:
            time.sleep(sleep)
        t_next = max(t_next + DT, now)

finally:
    cap.release()
    print("✅ סיימתי")
````

## 🏁 שלב 6: המשימה האמיתית - עם תנועה!

**⚠️⚠️⚠️ חשוב מאוד! ⚠️⚠️⚠️**

לפני ההרצה:
1. ✅ שימו את הרובוט על **הרצפה** (לא על שולחן!)
2. ✅ וודאו שיש **מקום פנוי** מסביב
3. ✅ הדפיסו/הציגו את הלוחית `TARGET_PLATE` על קיר/קופסה
4. ✅ הכוונו את הרובוט כך שהמצלמה רואה את הלוחית
5. ✅ הכינו את המקש 'q' לעצירת חירום!

**איך זה עובד:**
1. הרובוט מזהה לוחיות
2. קורא את המספרים
3. כשמוצא את המטרה → **נועל** עליה
4. מחשב זווית ומהירות
5. נוסע לעברה
6. עוצר כשמגיע (לוחית תופסת 60% מהמסך)

**פקודות:**
- **'q'** בחלון → עצירת חירום
- **ESC** → עצירת חירום

````python
# --- create robot ---
car = YB_Pcb_Car()

# --- Jupyter camera widget connection (replaces cv2.VideoCapture + cv2.imshow) ---
cap, show_frame = free_camera(index=0, width=640, height=480, use_v4l2=False, jpeg_quality=75)

print("="*50)
print(f"🎯 משימה: למצוא ולהגיע ל-{TARGET_PLATE}")
print(f"🚀 מהירות: {SPEED} PWM")
print(f"🛑 עצירה: כשלוחית תופסת {TARGET_RATIO*100:.0f}% מהמסך")
print("="*50)
print("⚠️ READY TO MOVE! עצירה: Interrupt לקורנל / Stop (אין q ב-Jupyter widget)")
print("="*50)
print()

last_infer = 0.0
detections = []
frame_count = 0
target_found = False
target_plate_text = None
yaw_filtered = 0.0
last_detection_time = 0.0
reached = False

# streaming loop parameters
TARGET_FPS = 20
DT = 1.0 / TARGET_FPS
DROP_GRABS = 2
t_next = time.time()

try:
    for i in range(2000):  # מקסימום 2000 פריימים
        # drop buffered frames to reduce lag
        for _ in range(DROP_GRABS):
            cap.grab()

        ret, frame = cap.read()
        if not ret:
            time.sleep(0.02)
            continue

        now = time.time()
        frame_count += 1
        fh, fw = frame.shape[:2]

        # AI
        if now - last_infer >= INFER_PERIOD_S:
            output = run_onnx_inference(frame)
            detections = decode_yolov8(output, fw, fh, CONF_THRES, IOU_THRES)
            last_infer = now

        yaw = 0.0
        L, R = 0, 0
        status = "Searching..."
        has_detection = False

        if reached:
            status = "Done!"
            car.Car_Stop()

        elif detections:
            best = max(detections, key=lambda d: d["confidence"])
            x1, y1, x2, y2 = best["bbox"]

            if passes_plate_filters((x1, y1, x2, y2)):
                has_detection = True
                last_detection_time = now

                cx, cy = bbox_center((x1, y1, x2, y2))
                yaw, _ = pixel_to_angles(cx, cy, K)
                yaw_filtered = (1 - YAW_FILTER_ALPHA) * yaw_filtered + YAW_FILTER_ALPHA * yaw

                bbox_w = x2 - x1

                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.circle(frame, (int(cx), int(cy)), 5, (0, 0, 255), -1)

                # OCR (רק אם עדיין לא מצאנו)
                if not target_found and frame_count % OCR_EVERY_N_FRAMES == 0:
                    crop = crop_plate(frame, (x1, y1, x2, y2))
                    plate = ocr_plate(crop)
                    if plate:
                        print(f"📋 קראתי: {plate}")
                        if plate == TARGET_PLATE:
                            target_found = True
                            target_plate_text = plate
                            print(f"\n🎯 LOCK ON TARGET: {TARGET_PLATE}!\n")

                # תנועה רק אם מצאנו את המטרה
                if True: target_found:
                    if bbox_w >= (fw * TARGET_RATIO):
                        status = "TARGET REACHED!"
                        car.Car_Stop()
                        reached = True
                        print("\n🏁🏁🏁 הגעתי למטרה! 🏁🏁🏁\n")
                    else:
                        L, R = compute_steering(yaw_filtered)
                        car.Car_Run(int(L), int(R))
                        status = f"Approaching (w={bbox_w}px)"
                else:
                    status = "Plate detected, waiting OCR"
                    car.Car_Stop()

        # אם איבדנו זיהוי
        if not has_detection:
            if (now - last_detection_time) > LOST_TIMEOUT_S:
                status = "Target lost"
                car.Car_Stop()
            else:
                # ממשיכים עם yaw אחרון
                if target_found and not reached:
                    L, R = compute_steering(yaw_filtered)
                    car.Car_Run(int(L), int(R))
                    status = "Tracking last heading..."
                else:
                    car.Car_Stop()

        vis = draw_hud(frame, yaw_filtered, status, L, R, target_plate_text)

        # Jupyter widget output (replaces cv2.imshow + cv2.waitKey)
        show_frame(vis)

        # pace output (prevents CPU spikes)
        sleep = t_next - time.time()
        if sleep > 0:
            time.sleep(sleep)
        t_next = max(t_next + DT, time.time())

finally:
    car.Car_Stop()
    cap.release()
    print("\n" + "="*50)
    if reached:
        print("✅ המשימה הושלמה בהצלחה!")
    else:
        print("⚠️ המשימה הופסקה")
    print("="*50)
````

## 📊 סיכום - כל הפרמטרים במקום אחד

### טבלת כיוונון מהיר:

| קטגוריה | פרמטר | ברירת מחדל | אם רוצים... |
|----------|--------|------------|-------------|
| **מהירות** | `SPEED` | 80 | איטי יותר → 50-60, מהיר יותר → 100-120 |
| **דיוק היגוי** | `YAW_DEADBAND` | 5.0 | דיוק גבוה → 2-3, נוח יותר → 8-10 |
| **תגובתיות** | `STEER_KP` | 0.020 | חד יותר → 0.030-0.040, עדין יותר → 0.010-0.015 |
| **מרחק עצירה** | `TARGET_RATIO` | 0.6 | קרוב יותר → 0.7-0.8, רחוק יותר → 0.4-0.5 |
| **רגישות AI** | `CONF_THRES` | 0.2 | מחמיר יותר → 0.4-0.6, מקל יותר → 0.1-0.15 |
| **תדירות OCR** | `OCR_EVERY_N_FRAMES` | 3 | מהיר יותר → 1-2, חוסך CPU → 5-7 |

## 🎓 מה למדנו בפרויקט הזה?

### מושגי ליבה:

1. **Computer Vision (ראייה ממוחשבת)**
   - זיהוי אובייקטים (YOLO)
   - עיבוד תמונה (CLAHE, threshold)
   - OCR (קריאת טקסט)

2. **Robotics (רובוטיקה)**
   - שליטה במנועים (PWM)
   - חישוב זוויות (geometry)
   - Path planning (תכנון נתיב)

3. **Control Theory (תורת הבקרה)**
   - בקר P (Proportional)
   - Deadband (איזור מת)
   - Filtering (סינון רעש)

4. **State Machines (מכונת מצבים)**
   - מחפש → מזהה → קורא → ננעל → ניגש → הגיע

### שרשרת החלטות:

```
תמונה
  ↓
AI: "יש לוחית ב-X,Y?"
  ↓
פילטר: "היא טובה?"
  ↓
OCR: "מה כתוב?"
  ↓
השוואה: "זו המטרה?"
  ↓
גיאומטריה: "איפה היא יחסית אלי?"
  ↓
בקר: "איזו זווית לפנות?"
  ↓
מנועים: "כמה PWM לכל גלגל?"
  ↓
תנועה!
```

   - שני רובוטים: אחד עם לוחית, השני רודף
   - הרודף צריך למצוא ולהגיע!
