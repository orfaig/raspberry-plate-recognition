**×¤×¨××˜×¨×™×:**
- `CONF_THRES = 0.2` - ×¡×£ ×‘×™×˜×—×•×Ÿ (20% = ××§×œ, 80% = ×§×¤×“× ×™)
- `IOU_THRES = 0.2` - ×œ×× ×™×¢×ª ×–×™×”×•×™×™× ×›×¤×•×œ×™×
- `INFER_PERIOD_S = 0.03` - ×›××” ×¤×¢××™× ×‘×©× ×™×™×” ×œ×”×¨×™×¥ ××ª ×”-AI (×›×œ 30ms)

````python
# × ×ª×™×‘ ×œ××•×“×œ
ONNX_PATH = "/home/pi/Yahboom_project/uveye/code/raspberry-plate-recognition/yolov8n-license_plate.onnx"

# ×¤×¨××˜×¨×™ ×–×™×”×•×™
CONF_THRES = 0.2      # ×¡×£ ×‘×™×˜×—×•×Ÿ - × ××•×š = ××–×”×” ×™×•×ª×¨ ××‘×œ ×™×•×ª×¨ ×˜×¢×•×™×•×ª
IOU_THRES = 0.2       # ×˜×™×¤×•×œ ×‘×–×™×”×•×™×™× ×—×•×¤×¤×™×
INFER_PERIOD_S = 0.03 # ×ª×“×™×¨×•×ª ×”×¨×¦×ª AI (33 ×¤×¢××™× ×‘×©× ×™×™×”)

# ×˜×¢×™× ×ª ×”××•×“×œ
onnxr.set_default_logger_severity(3)  # ×”×©×ª×§×ª ×”×•×“×¢×•×ª ××™×•×ª×¨×•×ª
session = onnxr.InferenceSession(ONNX_PATH, providers=["CPUExecutionProvider"])

# ×‘×“×™×§×ª ×§×œ×˜ ×”××•×“×œ
inp = session.get_inputs()[0]
print(f"âœ… ××•×“×œ YOLO × ×˜×¢×Ÿ!")
print(f"ğŸ“Š ×§×œ×˜: {inp.name}, ×¦×•×¨×”: {inp.shape}")
print(f"âš™ï¸ CONF_THRES={CONF_THRES}, IOU_THRES={IOU_THRES}")
````

## ğŸ” ×©×œ×‘ 3: ×–×™×”×•×™ ×œ×•×—×™×ª ×¤×©×•×˜ (×œ×œ× OCR)

× ×ª×—×™×œ ×¨×§ ×¢× ×”×–×™×”×•×™ - × ×¨××” ××™×¤×” ×”××•×“×œ ××•×¦× ×œ×•×—×™×•×ª.

**××™×š ×–×” ×¢×•×‘×“:**
1. ×§×¨×™××ª ×ª××•× ×” ××”××¦×œ××”
2. ×©×œ×™×—×ª ×”×ª××•× ×” ×œ××•×“×œ YOLO
3. ×”××•×“×œ ××—×–×™×¨: "××¦××ª×™ ×œ×•×—×™×ª ×‘-X, Y ×¢× ×¨×•×—×‘ W ×•×’×•×‘×” H"
4. × ×¦×™×™×¨ ×¨×™×‘×•×¢ ×™×¨×•×§ ×¡×‘×™×‘ ×”×œ×•×—×™×ª

**×˜×™×¤:** × ×¡×• ×¢× ×ª××•× ×ª ×œ×•×—×™×ª ×¢×œ ××¡×š ×”×˜×œ×¤×•×Ÿ ××• ×”×“×¤×¡×”.

````python
# ×©×—×¨×•×¨ ××¦×œ××” ×× ×ª×§×•×¢×”
free_camera()

# ×¤×ª×™×—×ª ××¦×œ××”
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

print("ğŸ” ××–×”×” ×œ×•×—×™×•×ª (×œ×œ× OCR)...")
print("×”×¨××• ×œ×•×—×™×ª ×¨×›×‘ (××¤×©×¨ ×¢×œ ××¡×š/× ×™×™×¨)")

last_infer = 0.0
detections = []

for i in range(300):
    ret, frame = cap.read()
    if not ret:
        continue
    
    now = time.time()
    
    # ×”×¨×¦×ª AI ×¨×§ ×›×œ INFER_PERIOD_S (×œ×—×¡×•×š CPU)
    if now - last_infer >= INFER_PERIOD_S:
        output = run_onnx_inference(frame)
        detections = decode_yolov8(
            output, frame.shape[1], frame.shape[0],
            CONF_THRES, IOU_THRES
        )
        last_infer = now
    
    vis = frame.copy()
    
    # ×¦×™×•×¨ ×›×œ ×”×œ×•×—×™×•×ª ×©× ××¦××•
    for det in detections:
        x1, y1, x2, y2 = det["bbox"]
        conf = det["confidence"]
        
        cv2.rectangle(vis, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(vis, f"{conf:.2f}", (x1, y1-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)
    
    # ××•× ×”
    cv2.putText(vis, f"Plates: {len(detections)}", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 255), 2)
    
    cv2.imshow("×–×™×”×•×™ ×œ×•×—×™×•×ª", vis)
    if cv2.waitKey(1) == 27:  # ESC
        break

cap.release()
cv2.destroyAllWindows()
print("âœ… ×¡×™×™××ª×™")
````

## ğŸ”¤ ×©×œ×‘ 4: ×”×•×¡×¤×ª OCR - ×§×¨×™××ª ×”××¡×¤×¨×™×

×¢×›×©×™×• × ×•×¡×™×£ ××ª ×”×©×œ×‘ ×”×©× ×™: ×§×¨×™××ª ×”×˜×§×¡×˜ ××”×œ×•×—×™×ª.

### ×ª×”×œ×™×š ×¢×™×‘×•×“ ×”×ª××•× ×” (×œ×©×™×¤×•×¨ ×”×§×¨×™××”):

1. **×’×–×™×¨×” (crop)** - ×—×•×ª×›×™× ×¨×§ ××ª ×”×œ×•×—×™×ª + ×©×•×œ×™×™×
2. **×”××¨×” ×œ×©×—×•×¨-×œ×‘×Ÿ (grayscale)**
3. **CLAHE** - ×©×™×¤×•×¨ × ×™×’×•×“×™×•×ª (×›××• "×‘×”×™×¨×•×ª ××•×˜×•××˜×™×ª")
4. **×”×’×“×œ×” (resize)** - x2.4 - OCR ×¢×•×‘×“ ×˜×•×‘ ×™×•×ª×¨ ×¢×œ ×ª××•× ×•×ª ×’×“×•×œ×•×ª
5. **×˜×©×˜×•×© ×§×œ (blur)** - ××¡×™×¨ ×¨×¢×©
6. **×¡×£ (threshold)** - ×©×—×•×¨-×œ×‘×Ÿ ×—×“ (×œ×œ× ××¤×•×¨)
7. **OCR** - ×¤×¢××™×™×: ×¢×œ ×”×¨×’×™×œ ×•×¢×œ ×”×”×¤×•×š (×©×—×•×¨â†”×œ×‘×Ÿ)

### ×¤×•×¨××˜ ×œ×•×—×™×ª ×™×©×¨××œ×™×ª:
- 7 ×¡×¤×¨×•×ª: `12-345-67`
- 8 ×¡×¤×¨×•×ª: `123-45-678`

````python
# ×”×’×“×¨×•×ª OCR
OCR_EVERY_N_FRAMES = 3  # OCR ×¨×§ ×›×œ 3 ×¤×¨×™×™××™× (×—×•×¡×š ×–××Ÿ)
MARGIN_X = 0.10         # ×©×•×œ×™×™× ××•×¤×§×™×™× (10%)
MARGIN_Y = 0.25         # ×©×•×œ×™×™× ×× ×›×™×™× (25%)

# CLAHE - ×œ×©×™×¤×•×¨ × ×™×’×•×“×™×•×ª
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))

def normalize_plate(raw_text):
    """×××™×¨ OCR ×’×•×œ××™ ×œ×¤×•×¨××˜ ×œ×•×—×™×ª"""
    # ×”×¡×¨×ª ×›×œ ××” ×©×œ× ×¡×¤×¨×”
    digits = re.sub(r"\D", "", raw_text or "")
    
    if len(digits) == 7:
        return f"{digits[:2]}-{digits[2:5]}-{digits[5:]}"
    elif len(digits) == 8:
        return f"{digits[:3]}-{digits[3:5]}-{digits[5:]}"
    return None

def crop_plate(frame, bbox):
    """×’×•×–×¨ ××ª ×”×œ×•×—×™×ª ×¢× ×©×•×œ×™×™×"""
    x1, y1, x2, y2 = bbox
    w, h = x2 - x1, y2 - y1
    mx = int(w * MARGIN_X)
    my = int(h * MARGIN_Y)
    
    return frame[
        max(0, y1 - my):min(frame.shape[0], y2 + my),
        max(0, x1 - mx):min(frame.shape[1], x2 + mx)
    ]

def ocr_plate(crop_img):
    """××‘×¦×¢ OCR ×¢×œ ×ª××•× ×ª ×œ×•×—×™×ª"""
    if crop_img is None or crop_img.size == 0:
        return None
    
    # ×¢×™×‘×•×“
    gray = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
    gray = clahe.apply(gray)
    gray = cv2.resize(gray, None, fx=2.4, fy=2.4, interpolation=cv2.INTER_CUBIC)
    gray = cv2.GaussianBlur(gray, (3, 3), 0)
    
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # × ×™×¡×™×•×Ÿ OCR ×¢×œ ×©×ª×™ ×’×¨×¡××•×ª (×¨×’×™×œ + ×”×¤×•×š)
    for img in (binary, 255 - binary):
        text = pytesseract.image_to_string(img, config=CONFIG_TESS)
        plate = normalize_plate(text)
        if plate:
            return plate
    
    return None

print("âœ… ×¤×•× ×§×¦×™×•×ª OCR ××•×›× ×•×ª!")
print(f"ğŸ“‹ ×¤×•×¨××˜×™×: XX-XXX-XX (7) ××• XXX-XX-XXX (8)")
````

## ğŸ¯ ×©×œ×‘ 5: ×–×™×”×•×™ + ×§×¨×™××” ××œ××”

×¢×›×©×™×• × ×©×œ×‘ ×”×›×œ: ×–×™×”×•×™ + OCR.

**××” ×™×•×¦×’:**
- ×¨×™×‘×•×¢ ×™×¨×•×§ ×¡×‘×™×‘ ×”×œ×•×—×™×ª
- ×”×˜×§×¡×˜ ×©× ×§×¨× (××• "----" ×× ×œ× ×”×¦×œ×™×—)
- ×¦×‘×¢ ×™×¨×•×§ ×× ×”×¦×œ×™×— ×œ×§×¨×•× ×¤×•×¨××˜ ×—×•×§×™

````python
free_camera()

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

print("ğŸ” ×–×™×”×•×™ + ×§×¨×™××” ××œ××”")
print("×”×¨××• ×œ×•×—×™×ª ×¨×›×‘ - ×”××¢×¨×›×ª ×ª×§×¨× ××ª ×”××¡×¤×¨\n")

last_infer = 0.0
detections = []
frame_count = 0
last_ocr_result = {}

for i in range(500):
    ret, frame = cap.read()
    if not ret:
        continue
    
    now = time.time()
    frame_count += 1
    
    # ×–×™×”×•×™
    if now - last_infer >= INFER_PERIOD_S:
        output = run_onnx_inference(frame)
        detections = decode_yolov8(
            output, frame.shape[1], frame.shape[0],
            CONF_THRES, IOU_THRES
        )
        last_infer = now
    
    vis = frame.copy()
    
    if detections:
        # × ×¢×‘×•×“ ×¢×œ ×”×œ×•×—×™×ª ×”×›×™ ×‘×˜×•×—×”
        best = max(detections, key=lambda d: d["confidence"])
        x1, y1, x2, y2 = best["bbox"]
        
        cv2.rectangle(vis, (x1, y1), (x2, y2), (0, 255, 0), 2)
        
        # OCR ×›×œ N ×¤×¨×™×™××™×
        if frame_count % OCR_EVERY_N_FRAMES == 0:
            crop = crop_plate(frame, (x1, y1, x2, y2))
            plate_text = ocr_plate(crop)
            
            if plate_text:
                last_ocr_result = {
                    "text": plate_text,
                    "time": now,
                    "bbox": (x1, y1, x2, y2)
                }
                print(f"ğŸ“‹ ×§×¨××ª×™: {plate_text}")
        
        # ×”×¦×’×ª ×ª×•×¦××” ××—×¨×•× ×” (×× ×§×™×™××ª ×•×¢×“×›× ×™×ª)
        if last_ocr_result and (now - last_ocr_result.get("time", 0)) < 2.0:
            txt = last_ocr_result["text"]
            cv2.putText(vis, f"Plate: {txt}", (10, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 3)
        else:
            cv2.putText(vis, "OCR: ----", (10, 40),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.0, (255, 255, 255), 3)
    
    cv2.imshow("×–×™×”×•×™ + OCR", vis)
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()
print("âœ… ×¡×™×™××ª×™")
````

## ğŸ¯ ×©×œ×‘ 6: ×—×™×¤×•×© ×œ×•×—×™×ª ×¡×¤×¦×™×¤×™×ª

×¢×›×©×™×• × ×•×¡×™×£ "××©×™××”": ×œ××¦×•× ×œ×•×—×™×ª ××¡×•×™××ª!

**××™×š ×–×” ×¢×•×‘×“:**
1. ×”×’×“×¨× ×• ×œ×•×—×™×ª ××˜×¨×”: `TARGET_PLATE = "91-179-18"`
2. ×”××¢×¨×›×ª ××–×”×” + ×§×•×¨××ª ×œ×•×—×™×•×ª
3. ×›×©××•×¦××ª ××ª ×”××˜×¨×” â†’ **LOCK!** (×œ× ××—×¤×©×ª ×™×•×ª×¨)
4. ×”×¦×’×” ×™×¨×•×§×” + ×”×•×“×¢×”

**×©×™××•×©:** ××©×—×§ "××¦× ××ª ×”×¨×›×‘" ××• ×‘×§×¨×ª ×›× ×™×¡×” ×œ×—× ×™×•×Ÿ.

````python
# ×œ×•×—×™×ª ××˜×¨×” (×©× ×• ×œ×¤×™ ×”×¦×•×¨×š!)
TARGET_PLATE = "91-179-18"

free_camera()

cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

print(f"ğŸ¯ ××—×¤×© ×œ×•×—×™×ª: {TARGET_PLATE}")
print("×”×¨××• ×œ×•×—×™×•×ª ×©×•× ×•×ª ×¢×“ ×©×ª××¦× ××ª ×”××˜×¨×”\n")

last_infer = 0.0
detections = []
frame_count = 0
target_found = False
target_data = None
last_guess = None

for i in range(1000):
    ret, frame = cap.read()
    if not ret:
        continue
    
    now = time.time()
    frame_count += 1
    
    # ×–×™×”×•×™
    if now - last_infer >= INFER_PERIOD_S:
        output = run_onnx_inference(frame)
        detections = decode_yolov8(
            output, frame.shape[1], frame.shape[0],
            CONF_THRES, IOU_THRES
        )
        last_infer = now
    
    vis = frame.copy()
    
    if detections:
        best = max(detections, key=lambda d: d["confidence"])
        x1, y1, x2, y2 = best["bbox"]
        
        # OCR ×¨×§ ×× ×¢×“×™×™×Ÿ ×œ× ××¦×× ×•
        if not target_found and frame_count % OCR_EVERY_N_FRAMES == 0:
            crop = crop_plate(frame, (x1, y1, x2, y2))
            plate_text = ocr_plate(crop)
            
            if plate_text:
                last_guess = plate_text
                print(f"ğŸ“‹ ×§×¨××ª×™: {plate_text}")
                
                # ×‘×“×™×§×” ××•×œ ××˜×¨×”
                if plate_text == TARGET_PLATE:
                    target_found = True
                    target_data = {
                        "text": plate_text,
                        "bbox": (x1, y1, x2, y2),
                        "time": now
                    }
                    print(f"\nğŸ‰ğŸ‰ğŸ‰ ××¦××ª×™ ××ª ×”××˜×¨×”: {TARGET_PLATE} ğŸ‰ğŸ‰ğŸ‰\n")
        
        # ×¦×™×•×¨
        color = (0, 255, 0) if target_found else (255, 255, 0)
        cv2.rectangle(vis, (x1, y1), (x2, y2), color, 3 if target_found else 2)
    
    # ×ª×¦×•×’×”
    if target_found:
        cv2.putText(vis, f"TARGET: {TARGET_PLATE}", (10, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.2, (0, 255, 0), 3)
        cv2.putText(vis, "*** FOUND ***", (10, 80),
                    cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 255, 0), 3)
    else:
        cv2.putText(vis, f"Looking for: {TARGET_PLATE}", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        if last_guess:
            cv2.putText(vis, f"Last: {last_guess}", (10, 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 0), 2)
    
    cv2.imshow("××—×¤×© ××˜×¨×”", vis)
    if cv2.waitKey(1) == 27:
        break

cap.release()
cv2.destroyAllWindows()

if target_found:
    print(f"âœ… ×¡×™×™××ª×™ - ××¦××ª×™ ××ª {TARGET_PLATE}!")
else:
    print("âš ï¸ ×¡×™×™××ª×™ - ×œ× ××¦××ª×™ ××ª ×”××˜×¨×”")
````
