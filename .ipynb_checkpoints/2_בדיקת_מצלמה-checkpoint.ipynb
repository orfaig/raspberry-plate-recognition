## 📚 שלב 1: ייבוא הספריות

**הספריות שנשתמש בהן:**
- `cv2` (OpenCV) - ספרייה לעבודה עם מצלמות ותמונות
- `Flask` - שרת אינטרנט קטן שישדר את התמונות לטלפון
- `threading` - כדי להריץ את השרת ברקע

**מושגים בסיסיים:**
- **Frame** = תמונה אחת מתוך הוידאו (כמו דף בספר מצויר)
- **FPS** = כמה תמונות בשנייה (30 FPS = 30 תמונות בשנייה = וידאו חלק)
- **Resolution** = גודל התמונה (640x480 = רוחב 640 פיקסלים, גובה 480)

````python
import os
os.environ["QT_QPA_PLATFORM"] = "xcb"  # הגדרה טכנית ל-Raspberry Pi

import cv2
import time
import threading
from flask import Flask, Response

print("✅ כל הספריות נטענו בהצלחה!")
````

---

## 🎥 שלב 2: פתיחת המצלמה (פשוט)

נתחיל עם הדרך הכי פשוטה - לפתוח מצלמה ולהציג תמונות חיות על המסך.

**פרמטרים חשובים:**
- `VideoCapture(0)` - המספר 0 אומר "המצלמה הראשונה"
- `CAP_PROP_FRAME_WIDTH` - רוחב התמונה
- `CAP_PROP_FRAME_HEIGHT` - גובה התמונה
- `CAP_PROP_FPS` - כמה תמונות בשנייה

**איך זה עובד:**
1. פותחים את המצלמה
2. בלולאה: קוראים תמונה אחת
3. מציגים אותה על המסך
4. חוזרים לשלב 2 (עד שלוחצים 'q' לעצירה)

````python
# פתיחת המצלמה
cap = cv2.VideoCapture(0)

# הגדרת איכות
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)   # רוחב 640 פיקסל
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)  # גובה 480 פיקסל
cap.set(cv2.CAP_PROP_FPS, 30)            # 30 תמונות לשנייה

if not cap.isOpened():
    print("❌ שגיאה: לא הצלחתי לפתוח את המצלמה")
else:
    print("✅ המצלמה פתוחה ומוכנה!")
    print("לחצו 'q' בחלון התמונה כדי לעצור")
    
    # לולאה - קריאת תמונות והצגה
    for i in range(100):  # 100 תמונות (כדי שלא ירוץ לנצח)
        ret, frame = cap.read()  # קריאת תמונה אחת
        
        if not ret:
            print("❌ לא הצלחתי לקרוא תמונה")
            break
        
        # הצגת התמונה בחלון
        cv2.imshow("המצלמה שלי", frame)
        
        # המתן קצר + בדיקה אם לחצו 'q'
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            print("עצרת את הצילום")
            break
    
    # סגירה מסודרת
    cap.release()
    cv2.destroyAllWindows()
    print("✅ המצלמה נסגרה")
````

## 📱 שלב 3: שידור לטלפון (Streaming)

עכשיו נוסיף שרת אינטרנט קטן שישדר את התמונות לטלפון.

**איך זה עובד בקצרה:**
1. השרת Flask רץ ברקע
2. הוא "מאזין" לכתובת `http://10.2.0.17:8080`
3. כשנפתח את הכתובת בטלפון, השרת שולח תמונות חדשות כל הזמן
4. זה נקרא **MJPEG stream** - כמו "תמונות רצות"

**מה קורה פה:**
- `@app.route("/")` = דף הבית (עם כפתורים)
- `@app.route("/video_feed")` = הוידאו עצמו
- `generate_stream()` = פונקציה שמייצרת תמונות אחת אחרי השנייה

**לא צריך להבין את כל הקוד!** זה רק "מתכון" שמשדר תמונות.

````python
# יצירת השרת
app = Flask(__name__)

# משתנים גלובליים
latest_jpeg = None  # התמונה האחרונה (בפורמט JPEG)
running = True      # דגל - האם להמשיך לרוץ?

# פונקציה שמייצרת stream של תמונות
def generate_stream():
    global latest_jpeg, running
    while running:
        if latest_jpeg is not None:
            # שליחת תמונה אחת (בפורמט MJPEG)
            yield (b"--frame\r\n"
                   b"Content-Type: image/jpeg\r\n\r\n" +
                   latest_jpeg + b"\r\n")
        else:
            time.sleep(0.02)  # ממתין אם אין תמונה

# דף הבית - HTML פשוט עם הוידאו
@app.route("/")
def index():
    return """
    <html>
      <head><title>מצלמת הרובוט</title></head>
      <body style="text-align:center; font-family:Arial;">
        <h1>🤖 מצלמת הרובוט שלי</h1>
        <img src="/video_feed" style="max-width:100%; border:2px solid blue;" />
        <p>
          <button onclick="location.reload()">רענן דף</button>
        </p>
      </body>
    </html>
    """

# נתיב הוידאו
@app.route("/video_feed")
def video_feed():
    return Response(generate_stream(),
                    mimetype="multipart/x-mixed-replace; boundary=frame")

print("✅ השרת מוכן (עדיין לא פועל)")
````

## 🚀 שלב 4: הרצת הכל ביחד

עכשיו נפעיל את הכל:
1. השרת ירוץ ברקע (thread נפרד)
2. המצלמה תתפוס תמונות
3. כל תמונה תשודר גם למסך וגם לטלפון

**איך לראות את הוידאו:**
- **במחשב:** חלון יקפוץ אוטומטית
- **בטלפון:** פתחו דפדפן וגשו ל: `http://10.2.0.17:8080`

**⚠️ חשוב:** הקוד הזה ירוץ עד שתלחצו 'q' בחלון!

````python
def run_flask():
    """מריץ את שרת Flask ברקע"""
    app.run(host="0.0.0.0", port=8080, debug=False, use_reloader=False)

def main():
    global latest_jpeg, running
    
    # הפעלת השרת ברקע
    flask_thread = threading.Thread(target=run_flask, daemon=True)
    flask_thread.start()
    print("🌐 השרת עלה! פתחו בטלפון: http://10.2.0.17:8080")
    
    # פתיחת המצלמה
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    cap.set(cv2.CAP_PROP_FPS, 30)
    
    if not cap.isOpened():
        print("❌ שגיאה: לא הצלחתי לפתוח מצלמה")
        running = False
        return
    
    print("📷 המצלמה פועלת! לחצו 'q' בחלון כדי לעצור")
    
    # לולאה ראשית
    while running:
        ret, frame = cap.read()
        
        if not ret:
            print("⚠️ לא קיבלתי תמונה מהמצלמה")
            time.sleep(0.1)
            continue
        
        # המרה לפורמט JPEG (לשידור לטלפון)
        ok, buffer = cv2.imencode(".jpg", frame)
        if ok:
            latest_jpeg = buffer.tobytes()
        
        # הצגה על המסך
        cv2.imshow("מצלמת הרובוט (לחץ q לעצירה)", frame)
        
        # בדיקת מקש
        key = cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            running = False
            break
    
    # סגירה
    cap.release()
    cv2.destroyAllWindows()
    print("✅ המצלמה נסגרה. השרת עדיין פועל ברקע.")

# הרצה!
if __name__ == "__main__":
    main()
````

## 🔧 ניסוי: שינוי רזולוציה ו-FPS

נסו לשנות את ההגדרות ולראות איך זה משפיע:

| פרמטר | ערך נמוך | ערך בינוני | ערך גבוה | השפעה |
|--------|----------|------------|----------|--------|
| **Width** | 320 | 640 | 1280 | רוחב התמונה - גדול יותר = איכות טובה יותר אבל איטי יותר |
| **Height** | 240 | 480 | 720 | גובה התמונה |
| **FPS** | 15 | 30 | 60 | תמונות בשנייה - גבוה יותר = חלק יותר אבל עוזר יותר CPU |

**דוגמה לרזולוציה נמוכה (מהיר):**

````python
# ניסוי: רזולוציה נמוכה
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 320)   # רוחב קטן
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 240)  # גובה קטן
cap.set(cv2.CAP_PROP_FPS, 15)            # 15 תמונות לשנייה

print("📷 מצלמה ברזולוציה נמוכה (320x240, 15 FPS)")
print("זה יהיה מהיר אבל פחות חד")

for i in range(50):
    ret, frame = cap.read()
    if ret:
        cv2.imshow("רזולוציה נמוכה", frame)
        if cv2.waitKey(1) == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
print("✅ סיימתי")
````

## 📊 סיכום - פקודות ופרמטרים

### פקודות מצלמה:

| פקודה | תיאור | פרמטרים |
|-------|--------|----------|
| `cv2.VideoCapture(0)` | פותח מצלמה | 0 = מצלמה ראשונה |
| `cap.set(property, value)` | שינוי הגדרות | ראו טבלה למטה |
| `cap.read()` | קריאת תמונה | מחזיר: (הצלחה?, תמונה) |
| `cv2.imshow(name, img)` | הצגת תמונה | שם החלון, התמונה |
| `cv2.waitKey(1)` | המתנה 1ms + בדיקת מקש | מחזיר: קוד המקש |
| `cap.release()` | סגירת המצלמה | - |
| `cv2.destroyAllWindows()` | סגירת כל החלונות | - |

### הגדרות מצלמה (properties):

| Property | ערך | תיאור |
|----------|-----|--------|
| `CAP_PROP_FRAME_WIDTH` | 320-1920 | רוחב פיקסלים |
| `CAP_PROP_FRAME_HEIGHT` | 240-1080 | גובה פיקסלים |
| `CAP_PROP_FPS` | 15-60 | תמונות לשנייה |
