## 📐 שלב 2: הבנת חישוב זוויות

**השאלה:** ה-QR נמצא בפיקסל (x=400, y=240). איזו זווית זה?

**התשובה:** תלוי בפרמטרי המצלמה!

### פרמטרי מצלמה (Camera Intrinsics):

- **fx, fy** = אורך מוקד (focal length) - כמה "זום" יש למצלמה
- **cx, cy** = מרכז התמונה (לרוב רוחב/2, גובה/2)

**דוגמה:**
- תמונה 640x480
- מרכז: (320, 240)
- QR ב-(400, 240) = 80 פיקסל ימינה
- עם fx=450 → זווית ≈ 10° ימינה

**הנוסחה (פשוטה):**
```python
yaw_deg = atan((x - cx) / fx) * (180/π)
```

**לא צריך לזכור את זה!** הפונקציה `pixel_to_angles` עושה את זה בשבילנו.

````python
# הגדרת פרמטרי מצלמה
FX = 450.0  # focal length X
FY = 600.0  # focal length Y
CX = 320.0  # center X (מרכז התמונה)
CY = 240.0  # center Y

# מטריצה K (Camera Intrinsics Matrix)
K = np.array([[FX,   0.0, CX],
              [0.0,  FY,  CY],
              [0.0,  0.0, 1.0]], dtype=np.float32)

print("📐 מטריצת המצלמה:")
print(K)
print()

# דוגמה: חישוב זווית
# נניח ש-QR נמצא ב-(400, 240)
x, y = 400, 240
yaw, pitch = pixel_to_angles(x, y, K)

print(f"QR בפיקסל ({x}, {y})")
print(f"→ Yaw (סיבוב אופקי): {yaw:.1f}°")
print(f"→ Pitch (סיבוב אנכי): {pitch:.1f}°")
print()
print("💡 Yaw חיובי = ימינה, Yaw שלילי = שמאלה")
````

## 🎮 שלב 3: מערכת היגוי (Steering)

**המטרה:** להמיר זווית (yaw) לפקודות גלגלים (left_pwm, right_pwm).

### עקרון הפעולה:

1. **אם yaw קטן (כמעט מרכז):**
   - נסע ישר קדימה: `left = right = 80`

2. **אם yaw חיובי (ימינה):**
   - גלגל שמאל מהיר יותר: `left = 100, right = 60`
   - הרובוט פונה ימינה

3. **אם yaw שלילי (שמאלה):**
   - גלגל ימין מהיר יותר: `left = 60, right = 100`
   - הרובוט פונה שמאלה

### פרמטרים לכיוונון:

- **SPEED** = מהירות בסיס (PWM)
- **YAW_DEADBAND** = "איזור מת" - אם |yaw| < זה, נסע ישר
- **STEER_KP** = מקדם היגוי - כמה להגיב לשגיאת זווית
- **STEER_LIMIT** = גבול תיקון מקסימלי

````python
# פרמטרי היגוי
SPEED = 80.0          # מהירות בסיס
YAW_DEADBAND = 5.0    # איזור מת (מעלות)
STEER_KP = 0.020      # מקדם P (proportional)
STEER_LIMIT = 0.25    # גבול תיקון (0.25 = 25%)

def compute_steering(yaw_deg):
    """מחשב PWM לגלגלים על בסיס זווית yaw"""
    
    # אם בתוך איזור המת - נסע ישר
    if abs(yaw_deg) <= YAW_DEADBAND:
        return SPEED, SPEED
    
    # חישוב תיקון היגוי
    steer = STEER_KP * yaw_deg
    steer = np.clip(steer, -STEER_LIMIT, STEER_LIMIT)
    
    # חישוב PWM
    left_pwm = SPEED * (1.0 + steer)
    right_pwm = SPEED * (1.0 - steer)
    
    # גבולות (PWM בטווח 0-100)
    left_pwm = np.clip(left_pwm, 0, 100)
    right_pwm = np.clip(right_pwm, 0, 100)
    
    return left_pwm, right_pwm

# דוגמאות
print("🎮 דוגמאות היגוי:\n")

for yaw in [-15, -5, 0, 5, 15]:
    L, R = compute_steering(yaw)
    direction = "⬅️ שמאל" if yaw < -YAW_DEADBAND else "➡️ ימין" if yaw > YAW_DEADBAND else "⬆️ ישר"
    print(f"Yaw={yaw:+3.0f}° → L={L:.0f} R={R:.0f} {direction}")
````

## 🚀 שלב 4: מעקב אחרי QR (ללא תנועה)

נתחיל בלי להזיז את הרובוט - רק נראה מה הוא "חושב".

**מה יוצג על המסך:**
- חץ שמראה לאן לפנות
- Yaw (זווית)
- PWM של הגלגלים (כמה לכל אחד)

**התרגיל:** הזיזו את ה-QR ותראו איך החץ והמספרים משתנים.

````python
# פרמטרי QR (מ-notebook 3)
QR_MIN_W = 35
QR_MIN_H = 35
QR_MIN_ASPECT = 0.65
QR_MAX_ASPECT = 1.35

def passes_filters(points):
    xs = points[:, 0]
    ys = points[:, 1]
    x1, y1 = int(np.min(xs)), int(np.min(ys))
    x2, y2 = int(np.max(xs)), int(np.max(ys))
    width = x2 - x1
    height = y2 - y1
    aspect = width / max(1, height)
    size_ok = (width >= QR_MIN_W) and (height >= QR_MIN_H)
    aspect_ok = (QR_MIN_ASPECT <= aspect <= QR_MAX_ASPECT)
    return (size_ok and aspect_ok), x1, y1, x2, y2

def draw_hud(frame, yaw, left_pwm, right_pwm):
    """מצייר HUD (תצוגת מידע) על הפריים"""
    h, w = frame.shape[:2]
    
    # רקע שקוף למידע
    overlay = frame.copy()
    cv2.rectangle(overlay, (0, 0), (w, 100), (0, 0, 0), -1)
    frame = cv2.addWeighted(overlay, 0.35, frame, 0.65, 0)
    
    # חץ כיוון
    cx, cy = w // 2, 70
    arrow_len = 50
    yaw_clipped = np.clip(yaw, -45, 45)
    theta = np.radians(yaw_clipped)
    ex = int(cx + arrow_len * np.sin(theta))
    ey = int(cy - arrow_len * np.cos(theta))
    cv2.arrowedLine(frame, (cx, cy), (ex, ey), (0, 255, 0), 4, tipLength=0.3)
    
    # טקסט
    cv2.putText(frame, f"Yaw: {yaw:.1f} deg", (10, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)
    cv2.putText(frame, f"L:{left_pwm:.0f} R:{right_pwm:.0f}", (w-150, 30),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 255), 2)
    
    return frame

# הרצה
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

detector = cv2.QRCodeDetector()

print("🎯 מעקב אחרי QR (ללא תנועה)")
print("הזיזו QR ותראו איך הרובוט מגיב")

for i in range(300):
    ret, frame = cap.read()
    if not ret:
        continue
    
    text, points, _ = detector.detectAndDecode(frame)
    
    yaw = 0.0
    left_pwm = 0.0
    right_pwm = 0.0
    
    if points is not None:
        ok, x1, y1, x2, y2 = passes_filters(points[0])
        
        if ok:
            # חישוב מרכז
            cx = (x1 + x2) / 2.0
            cy = (y1 + y2) / 2.0
            
            # חישוב זווית
            yaw, _ = pixel_to_angles(cx, cy, K)
            
            # חישוב היגוי
            left_pwm, right_pwm = compute_steering(yaw)
            
            # ציור bbox
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.circle(frame, (int(cx), int(cy)), 5, (0, 0, 255), -1)
    
    # HUD
    frame = draw_hud(frame, yaw, left_pwm, right_pwm)
    
    cv2.imshow("מעקב QR", frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print("✅ סיימתי")
````

## 🏁 שלב 5: רובוט עוקב אמיתי (עם תנועה!)

עכשיו נחבר את הכל - הרובוט באמת יזוז!

### לוגיקה:

1. **מצא QR** → אם לא מוצא, עצור
2. **בדוק פילטרים** → אם לא עובר, עצור
3. **חשב yaw** → לאן לפנות?
4. **חשב PWM** → כמה לכל גלגל?
5. **נסע!** → `car.Car_Run(left, right)`
6. **בדוק הגעה** → אם ה-QR גדול מספיק (>60% מהמסך), עצור!

### תנאי עצירה:

```python
TARGET_RATIO = 0.60  # 60% מרוחב המסך
if bbox_width >= (frame_width * TARGET_RATIO):
    print("🎉 הגענו למטרה!")
    car.Car_Stop()
```

**⚠️ לפני ההרצה:**
1. שימו את הרובוט על הרצפה (לא על שולחן!)
2. הדביקו QR על קיר או קופסה
3. הכוונו את הרובוט כך שהמצלמה רואה את ה-QR
4. הריצו - הרובוט יתחיל לנוע!

````python
# פרמטרים
SPEED = 70.0          # מהירות (אפשר להתחיל נמוך!)
YAW_DEADBAND = 5.0
STEER_KP = 0.020
STEER_LIMIT = 0.25
TARGET_RATIO = 0.60   # עצור כש-QR תופס 60% מהמסך

# יצירת רובוט
car = YB_Pcb_Car()

# פתיחת מצלמה
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

detector = cv2.QRCodeDetector()

print("🤖 רובוט עוקב QR!")
print(f"מהירות: {SPEED}, תנאי עצירה: {TARGET_RATIO*100:.0f}% מהמסך")
print("⚠️ ודאו שהרובוט על הרצפה!")
print("לחצו 'q' לעצירת חירום\n")

reached = False

try:
    for i in range(1000):  # מקסימום 1000 פריימים
        ret, frame = cap.read()
        if not ret:
            continue
        
        h, w = frame.shape[:2]
        text, points, _ = detector.detectAndDecode(frame)
        
        status = "מחפש QR..."
        yaw = 0.0
        left_pwm = 0.0
        right_pwm = 0.0
        
        if reached:
            status = "🎉 הגעתי!"
            car.Car_Stop()
        elif points is not None:
            ok, x1, y1, x2, y2 = passes_filters(points[0])
            
            if ok:
                bbox_w = x2 - x1
                
                # בדיקת הגעה
                if bbox_w >= (w * TARGET_RATIO):
                    status = "🎯 TARGET REACHED!"
                    car.Car_Stop()
                    reached = True
                    print("\n🎉 הגעתי למטרה!")
                else:
                    # חישוב מרכז וזווית
                    cx = (x1 + x2) / 2.0
                    cy = (y1 + y2) / 2.0
                    yaw, _ = pixel_to_angles(cx, cy, K)
                    
                    # חישוב היגוי
                    left_pwm, right_pwm = compute_steering(yaw)
                    
                    # תנועה!
                    car.Car_Run(int(left_pwm), int(right_pwm))
                    status = f"נוסע (w={bbox_w}px)"
                
                # ציור
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.circle(frame, (int(cx), int(cy)), 5, (0, 0, 255), -1)
            else:
                status = "QR לא עובר פילטרים"
                car.Car_Stop()
        else:
            car.Car_Stop()
        
        # HUD
        frame = draw_hud(frame, yaw, left_pwm, right_pwm)
        cv2.putText(frame, status, (10, 60),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
        
        cv2.imshow("רובוט עוקב", frame)
        if cv2.waitKey(1) == ord('q'):
            print("\n⚠️ עצירת חירום!")
            break

finally:
    car.Car_Stop()
    cap.release()
    cv2.destroyAllWindows()
    print("✅ הרובוט עצר")
````

## 🔧 ניסוי כיוונון: השפעת הפרמטרים

שנו כל פרמטר בנפרד ותראו איך זה משפיע על ההתנהגות:

### 1. מהירות (`SPEED`)

| ערך | התנהגות |
|-----|----------|
| 40-60 | איטי ויציב, קל לכוונן |
| 70-90 | בינוני, איזון טוב |
| 100-120 | מהיר, אבל יכול "לעבור" את המטרה |

### 2. איזור מת (`YAW_DEADBAND`)

| ערך | התנהגות |
|-----|----------|
| 2-3 | רגיש מאוד, מתקן כל זווית קטנה → עלול להתנדנד |
| 5-7 | איזון טוב |
| 10-15 | מקל, נוסע ישר גם כשלא מכוון בדיוק → יכול לפספס |

### 3. מקדם היגוי (`STEER_KP`)

| ערך | התנהגות |
|-----|----------|
| 0.010 | סיבובים עדינים, איטיים להגיב |
| 0.020 | סטנדרטי |
| 0.040 | סיבובים חדים, עלול להתנדנד |

### 4. יחס מטרה (`TARGET_RATIO`)

| ערך | התנהגות |
|-----|----------|
| 0.40 | עוצר רחוק מה-QR |
| 0.60 | סטנדרטי - מרחק בינוני |
| 0.80 | עוצר קרוב מאוד (כמעט "מתנגש") |

## 📊 סיכום - מושגים ופקודות

### מושגים חדשים:

| מושג | הסבר | דוגמה |
|------|-------|-------|
| **Yaw** | זווית אופקית (שמאל/ימין) | +10° = ימינה, -10° = שמאלה |
| **Pitch** | זווית אנכית (למעלה/למטה) | +5° = למעלה |
| **PWM** | אות שליטה במנוע (0-255 או 0-100) | 80 = מהירות בינונית |
| **Deadband** | איזור שבו לא מתקנים | אם |yaw| < 5° → נסע ישר |
| **Proportional (P)** | תיקון יחסי לשגיאה | שגיאה גדולה → תיקון גדול |
| **Path Planning** | תכנון נתיב | "פנה 10° ונסע קדימה" |

### פונקציות עיקריות:

```python
# חישוב זווית
yaw, pitch = pixel_to_angles(x, y, K)

# חישוב היגוי
left_pwm, right_pwm = compute_steering(yaw)

# תנועה
car.Car_Run(left_pwm, right_pwm)
car.Car_Stop()
```
